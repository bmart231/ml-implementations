# Scratch Implementation

## Single Nueron, forward pass only

```python
import numpy as np

# inputs (3 features)
x = np.array([1.0, 2.0, 3.0])

# weights and bias for one neuron
w = np.array([0.2, 0.8, -0.5])
b = 2.0

# forward pass: weighted sum + bias
z = np.dot(x, w) + b
print(z)
```

## Step 2: Add an activation function, then make it a layer (multiple neurons).

```python
def relu(z):
    return np.maximum(z, 0)

a = relu(z)
print(a)
```

```python
# same input as before
x = np.array([1.0, 2.0, 3.0])

# now we have 3 neurons, so 3 weight vectors -> a weight MATRIX (3 neurons x 3 inputs)
W = np.array([
    [0.2, 0.8, -0.5],
    [0.5, -0.91, 0.26],
    [-0.26, -0.27, 0.17]
])

# 3 biases, one per neuron
b = np.array([2.0, 3.0, 0.5])

# forward pass for the whole layer at once
z = np.dot(W, x) + b
a = relu(z)

print("z:", z)
print("a:", a)
```
